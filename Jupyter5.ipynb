```python
import pandas as pd
import numpy as np
```

```python
link = "https://www.czso.cz/documents/10180/32853391/300002164403.xlsx/1fed9fb2-4dbf-4b0f-bfbe-4c2d3202a18f?version=1.1"
pd.read_excel(link)
data = pd.read_excel(link, header = None, skiprows = 6)
data.head(5)
```

```python
col1 = data.iloc[0].fillna(method = 'ffill')
col2 = data.iloc[2].fillna(method = 'ffill').str.split(pat='/', n = 0).str.get(0).fillna("")
col3 = data.iloc[3].fillna("")
columns = [col1, col2, col3]

columns = list(zip(*columns))
data.columns = columns
data = data.drop(data.index[[0,1,2]])
data = data.replace(".", "NaN")
data = data.replace("Lucembursko1)", "Lucembursko")
data.head(3)
```

```python
data = data.drop(data.columns[[-1]], axis=1)
data = data.replace("2)", "")
data = data.drop(data.index[[0, 1, -1,-2, -3]])
data.columns = pd.MultiIndex.from_tuples(data.columns, names=["", "", ""])
data.fillna("")
```
